## Aula 19 - Técnicas de Inpaint

Começamos importando as bibliotecas necessárias:

```python
%matplotlib inline
import cv2 # OpenCV
import matplotlib.pyplot as plt # Matplotlib
import numpy as np # Numpy
```

Precisamos de dois arquivos para realizar o inpaint:
- A imagem que queremos tratar;
- A máscara.

```python
img = plt.imread('inpainted.jpg')
plt.imshow(img)
```

```python
mascara = cv2.imread('mascara.jpg', 0)
plt.imshow(mascara, cmap='gray')
```

O inpaint é feito usando o método cv2.inpaint(), passando como argumento:
1. A imagem;
2. A máscara;
3. A vizinhança para fazer o inpaint. Se as regiões forem finas, menores valores reduzem o borramento;
4. As flags que definem o tipo de INPAINT, podendo ser INPAINT_NS para Navier-Stokes ou INPAINT_TELEA para fast-marching.

```python
inpainted = cv2.inpaint(img, mascara, 3, cv2.INPAINT_TELEA)
plt.imshow(inpainted)
```

Usamos o mesmo pipeline do último notebook para detectar e contar as faces:

```python
inpainted = cv2.inpaint(img, mascara, 3, cv2.INPAINT_NS)
plt.imshow(inpainted)
```

Qual dos dois é mais rápido?

```python
%timeit cv2.inpaint(img, mascara, 3, cv2.INPAINT_TELEA)
%timeit cv2.inpaint(img, mascara, 3, cv2.INPAINT_NS)
```

Junto com o repositório está um projeto de terceiros para carregar uma imagem e fazer o inpaint em tempo real. Após executar, basta desenhar na imagem e apertar **n** para ver o resultado com cv2.INPAINT_NS ou **t** para ver o resultado com cv2.INPAINT_TELEA.
