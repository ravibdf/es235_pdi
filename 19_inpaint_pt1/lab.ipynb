Precisamos de dois arquivos para realizar o inpaint:

- A máscara.

```python
img = plt.imread('inpainted.jpg')
plt.imshow(img)
```

```python
mascara = cv2.imread('mascara.jpg', 0)
plt.imshow(mascara, cmap='gray')
```

O inpaint é feito usando o método [cv2.inpaint()](https://docs.opencv.org/4.5.2/d7/d8b/group__photo__inpaint.html#gaedd30dfa0214fec4c88138b51d678085), passando como argumento:

1. A imagem;
2. A máscara;
3. A vizinhança para fazer o inpaint. Se as regiões forem finas, menores valores reduzem o borramento;
4. As flags que definem o tipo de INPAINT, podendo ser INPAINT_NS para Navier-Stokes ou INPAINT_TELEA para fast-marching.

Qual dos dois é mais rápido?

```python
%timeit cv2.inpaint(img, mascara, 3, cv2.INPAINT_TELEA)
%timeit cv2.inpaint(img, mascara, 3, cv2.INPAINT_NS)
```
