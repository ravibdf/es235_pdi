### Remoção de ruído

E agora, vamos carregar a imagem da Lenna:

```python
img = load_image('lenna.jpg', 256)
imageplot(img, 'Original image')
```

Adicionando ruído...

```python
img_r = img + .08*random.standard_normal(img.shape)
```

```python
imageplot(clamp(img_r), 'Ruidosa, SNR=' + str(snr(img,img_r)) )
```

Precisamos definir o limiar que irá separar o que é ruído do que é imagem:

```python
def limiar(u,t):
    return u*(abs(u)>t)

x = linspace(-3,3,1000)
plot(x, limiar(x,1))
axis('equal');
```

Alguns parâmetros para a transformada inversa:

```python
h = [0, .482962913145, .836516303738, .224143868042, -.129409522551]
h = h/norm(h)
Jmin = 2
```

E agora, podemos calcular os coeficientes do ruído da imagem:

```python
a = perform_wavortho_transf(img_r,Jmin,+1,h)
plot_wavelet(a,Jmin);
```

E os coeficientes já com o limiar aplicado:

```python
aT = limiar(a,3*.08)
plot_wavelet(aT,Jmin);
```

```python
img_n = perform_wavortho_transf(aT,Jmin,-1,h)
imageplot(clamp(img_n), 'Hard, SNR=' + str(snr(img,img_n)) )
```

### Compressão de imagens

```python
from mahotas.thresholding import soft_threshold
img = img[:256,:256]
# Show the data:
print("Quantidade de zeros na imagem original: {0}".format(np.mean(img==0)))
plt.imshow(img)
plt.show()
```

Usaremos o Wavelet chamado de [Daubechies wavelet](https://en.wikipedia.org/wiki/Daubechies_wavelet) e vamos descartar os bits de menor ordem:

```python
img.shape
```

```python
import numpy as np
import mahotas
import mahotas.demos

from mahotas.thresholding import soft_threshold
from matplotlib import pyplot as plt
from os import path
f = mahotas.demos.load('luispedro', as_grey=True)

f = f[:256,:256]
print(f.shape)
plt.gray()
# Show the data:
print("Fraction of zeros in original image: {0}".format(np.mean(f==0)))
plt.imshow(f)
plt.show()
```

```python
# Transform using D8 Wavelet to obtain transformed image t:
t = mahotas.daubechies(f,'D8')

# Discard low-order bits:
t /= 8
t = t.astype(np.int8)
print("Fração de zeros na transformada - pós divisão: {0}".format(np.mean(t==0)))
plt.imshow(t)
plt.show()
```

Temos 60% de zeros na imagem. Vamos usá-la para reconstruir a imagem original:

```python
r = mahotas.idaubechies(t, 'D8')
plt.imshow(r)
plt.show()
```

### Obtenção de características

```python
import pywt.data
original = pywt.data.camera()

titles = ['Aproximação', ' Detalhamento horizontal',
          'Detalhamento vertical', 'Detalhamento diagonal']
coeffs2 = pywt.dwt2(original, 'bior1.3')
LL, (LH, HL, HH) = coeffs2
fig = plt.figure(figsize=(12, 3))
for i, a in enumerate([LL, LH, HL, HH]):
    ax = fig.add_subplot(1, 4, i + 1)
    ax.imshow(a, interpolation="nearest", cmap=plt.cm.gray)
    ax.set_title(titles[i], fontsize=10)
    ax.set_xticks([])
    ax.set_yticks([])

fig.tight_layout()
plt.show()
```
